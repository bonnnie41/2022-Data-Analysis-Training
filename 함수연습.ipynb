### 산토끼

```python
lyric = '''
        산토끼 토끼야. 어디를 가느냐. 깡충깡충 뛰면서. 어디를 가느냐.
        산 고개 고개를. 나 혼자 넘어서. 토실토실 알밤을. 주워 올 테야
        '''
```

```python
def lyric_in(lylic,n):
    count = 0
    for i in lyric:
        if i == n:
            count += 1
    return count
```

```python
lyric_in(lyric,"토")
```

### 기온이 0ºC보다 높으면 ‘아이스 아메리카노’를 출력하고, 그렇지 않은 경우에는 ‘따뜻한 아메리카노’를 출력하는 함수를 만들어 보세요.

```python
def coffee(temp):
  if temp > 0:
    print("아이스 아메리카노")
  else:
    print("따뜻한 아메리카노")
```

```python
coffee(-1)
```

### 사용자로부터 키와 성별을 입력 받아서 권장 체중을 화면에 출력하는 '함수'를 작성하세요.
권장 체중은 다음과 같이 계산합니다.
-      남성 권장 체중 = (키 – 100)
-      여성 권장 체중 = (키 – 100) * 0.9

```python
def recommend(height,gender):
  if gender == "여성":
    print("여성권장체중 :", (height-100)*0.9)
  elif gender == "남성":
    print("남성권장체중 :", height-100)
  else: 
    print("올바른 성별값을 입력하시오")
```

```python
recommend(160,"여성")
```
